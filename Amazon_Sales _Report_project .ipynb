# Fulfillment Analysis

```python
df.drop('Customer-ID',axis=1,inplace=True)
df
```

```python
df.info()
```

```python
# Customer id
df['Customer ID'] = ['CUST' + str(i) for i in range(1, len(df) + 1)]
df.info()
# print(df.head())
```

```python
df['Total Sales']= df['Sales'].sum()
df.info()
```

```python
# Total sales per customer
customer_sales = df.groupby('Customer ID')['Sales'].sum().reset_index()
customer_sales.columns = ['Customer ID', 'Total Sales']

# Average order value
aov = df.groupby('Customer ID')['Sales'].mean().reset_index()
aov.columns = ['Customer ID', 'Average Order Value']

# Purchase frequency
frequency = df['Customer ID'].value_counts().reset_index()
frequency.columns = ['Customer ID', 'Purchase Frequency']

# Recency (days since last purchase)
df['Date'] = pd.to_datetime(df['Date'])  # Ensure the date column is datetime
recency = df.groupby('Customer ID')['Date'].max().reset_index()
recency['Recency (Days)'] = (pd.Timestamp.today() - recency['Date']).dt.days
recency = recency[['Customer ID', 'Recency (Days)']]
```

```python
# Merge all metrics into one DataFrame
customer_analysis = customer_sales.merge(aov, on='Customer ID') \
                                  .merge(frequency, on='Customer ID') \
                                  .merge(recency, on='Customer ID')

print(customer_analysis.head())
```

```python
# Segment customers based on Total Sales
customer_analysis['Sales Segment'] = pd.qcut(customer_analysis['Total Sales'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Segment customers based on Recency
customer_analysis['Recency Segment'] = pd.qcut(customer_analysis['Recency (Days)'], q=4, labels=['Recent', 'Moderately Recent', 'Old', 'Very Old'])

# Segment customers based on Frequency
# customer_analysis['Frequency Segment'] = pd.qcut(customer_analysis['Purchase Frequency'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
top_customers = customer_analysis.sort_values('Total Sales', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='Customer ID', y='Total Sales', data=top_customers, palette='viridis')
plt.title('Top 10 Customers by Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
customer_analysis['Sales Segment'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Customer Sales Segments')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.show()
```
